```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.utils import resample
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel(r"D:\Data Science Course\Jupyter\spam ham emails.xlsx")
```

```python
X = df["Email"]
Y = df["Label"]
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split (X, Y, random_state=354 , train_size=0.75)
```

```python
vec = CountVectorizer()
```

```python
X_tr = vec.fit_transform(X_tr).toarray()
```

```python
X_tr
```

```python
X_te = vec.transform(X_te).toarray()
```

```python
X_te
```

```python
model = MultinomialNB()
```

```python
model.fit(X_tr,Y_tr)
```

```python
model.score(X_tr, Y_tr)
```

```python
Y_pr = model.predict(X_te)
```
