```python
dummy = np.array(["A","B","C","D","E","F", "G"])
```

```python
data = {
    "Transaction_ID":[i for i in range (1,11)],
    "Items": [list(np.random.choice(dummy, np.random.randint(3,5))) for i in range(10)]
}
```
