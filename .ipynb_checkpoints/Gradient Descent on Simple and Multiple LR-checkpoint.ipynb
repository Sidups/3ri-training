```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=30, n_features=1, noise=15)
```

```python
lr = 0.01
m = 0
b = 0

plt.scatter(X, y)

for i in range(40):
    b_slope = -2*np.sum(y - m*X.ravel() - b)
    m_slope = -2*np.sum((y - m*X.ravel() - b)*X.ravel())

    b = b - (lr*b_slope)
    m = m - (lr*m_slope)

    y_pr = m*X.ravel() + b
    plt.plot(X.ravel(), y_pr)
plt.show()
```

```python
lr = 0.01
m = np.array([0, 0])
b = 0


for i in range(2000):
    b_slope = -2*np.mean(y - np.dot(m, X.T) - b)
    m_slope = -2*np.mean(X.T*(y - np.dot(m, X.T) - b), axis=1)

    b = b - (lr*b_slope)
    m = m - (lr*m_slope)
    
print("Slope",m, "Intercept",b)
```
