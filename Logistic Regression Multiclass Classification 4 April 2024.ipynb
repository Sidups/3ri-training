### We will short down the names

```python
df = pd.DataFrame(iris.data, columns=['sep_len',  'sep_wid',  'pet_len',  'pet_wid'])
```

```python
df
```

### Target is missing. We will add target.

```python
## Method 1.
df["target"]=iris.target
df.head()
```

```python
# Method 2.
# We will add names of target and for that we need to map them.
```

```python
df["target"] = list(map(lambda x:"Setosa" if x==0 else ("Versicolor" if x==1 else "Verginica"), iris.target))
```

```python
sns.scatterplot(df, x="sep_len", y="sep_wid", hue="target", palette="rocket")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

### We can see setosa seperatly but versocolor and verginica is mixe hence we will check with petal length and width.

```python
sns.scatterplot(df, x="pet_len", y="pet_wid", hue="target", palette="rocket")
plt.title= ("Petal Lenght vd Petal Width")
plt.show()
```

### Now we can see deferntials. Now we will check all possible options for classification.

```python
sns.pairplot(df, hue="target", palette="rocket")
```

```python
X = df.drop("target", axis=1)
Y = df["target"]
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split(X, Y, train_size=0.75, random_state=368)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_tr, Y_tr)
```

```python
model.score(X_tr,Y_tr)
```

```python
Y_pr = model.predict (X_te)
```

```python
accuracy_score(Y_te, Y_pr)
```

```python
confusion_matrix(Y_te, Y_pr)
```

```python
sns.heatmap(confusion_matrix(Y_te, Y_pr), annot=True)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
print(classification_report(Y_te, Y_pr))
```
